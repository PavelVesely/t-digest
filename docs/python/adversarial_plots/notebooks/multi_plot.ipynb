```python
import os
import pandas as pd
import matplotlib.pyplot as plt
```

```python
results_location  = '../../../../data/results/'
```

```python
files = os.listdir(results_location)
```

```python
filepath = results_location + files[0]
```

```python
with open(filepath) as f:
    df = pd.read_csv(f, header=0, sep=';')
```

```python
x_axis = df.columns[0]
```

```python
for col in df.columns[1:-1]:
    plt.plot(df[x_axis], df[col], label=col)
    
plt.legend()
plt.show()
```
